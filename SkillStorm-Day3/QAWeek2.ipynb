```python
df = pd.DataFrame(df)


def plot_sales_distribution():
    # Return unique genres
    genres = df['Genre'].unique()
    # Iterate over each unique genre, generating total sum for each
    genre_sales = {genre: df[df['Genre'] == genre]['Global_Sales'].sum() for genre in genres}
    
    plt.figure(figsize=(10, 6))
    plt.bar(genre_sales.keys(), genre_sales.values(), color='skyblue')
    plt.xlabel('Genre')
    plt.ylabel('Global Sales (in millions)')
    plt.title('Global Sales Distribution by Genre')
    plt.xticks(rotation=45)
    plt.show()




def plot_sales_by_console():
    # Group sales by platform
    console_sales = df.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False)
    
    plt.figure(figsize=(12, 6))
    console_sales.plot(kind='bar', color='lightblue')
    plt.xlabel('Console')
    plt.ylabel('Global Sales (in millions)')
    plt.title('Global Sales by Console')
    plt.xticks(rotation=45)
    plt.show()


def plot_sales_by_top_publishers():
    # Group sales by top 10 publishers, as there are 578 unique publishers 
    publisher_sales = df.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False).head(10)
    
    plt.figure(figsize=(12, 6))
    publisher_sales.plot(kind='bar', color='lightgreen')
    plt.xlabel('Publisher')
    plt.ylabel('Global Sales (in millions)')
    plt.title('Top 10 Publishers by Global Sales')
    plt.xticks(rotation=45)
    plt.show()

# Function to show basic statistics
def show_statistics():
    print("Basic Statistics:")
    print(df.describe())

# Plot sales distribution by genre
plot_sales_distribution()

# Plot sales by console
plot_sales_by_console()

# Plot sales by publisher
plot_sales_by_top_publishers()

# Show basic statistics
show_statistics()
```

```python
def plot_nintendo_sales_per_year_with_fit(df):
   
    # Sort by publisher "Nintendo"
    nintendo_sales = df[df['Publisher'] == 'Nintendo']
    # Sum of yearly sales of Nintendo Sales 
    yearly_sales = nintendo_sales.groupby('Year')['Global_Sales'].sum().reset_index()
    # Sort yearly sales
    yearly_sales = yearly_sales.sort_values('Year')

    plt.figure(figsize=(14, 8))
    # Year as X axis
    # Global_Sales as Y axis
    plt.plot(yearly_sales['Year'], yearly_sales['Global_Sales'], color='blue', label='Global Sales', marker='o')

    
    x = yearly_sales['Year']
    y = yearly_sales['Global_Sales']
    # Using numpy, create line of best fit
    # Linear fit of 1
    # Returns array of Int [m, b] with line of best fit
    fit = np.polyfit(x, y, deg=1)
    
    # Using same plot, place line of best fit
    # Represents mx + b
    plt.plot(x, fit[0] * x + fit[1], color='red', linestyle='--', label='Best Fit Line')

    
    plt.title('Nintendo Sales per Year with Line of Best Fit')
    plt.xlabel('Year')
    plt.ylabel('Global Sales (in millions)')
    plt.legend()
    plt.tight_layout()
    plt.show()
    

plot_nintendo_sales_per_year_with_fit(df)
```

```python
def plot_top_three_publishers():
    # Create DF with filtered information, where publisher is among the top three publishers
    df_filtered = df[df['Publisher'].isin(['Nintendo', 'Electronic Arts', 'Activision'])]

    # Use pivot table to summarize 
    # Row = Year
    # Column = Publisher
    # Use each companies global sales as column value, 0 if null
    sales_per_year_publisher = df_filtered.pivot_table(index='Year', columns='Publisher', values='Global_Sales', aggfunc='sum')

    # Fill missing values with with 0
    sales_per_year_publisher = sales_per_year_publisher.fillna(0)

    # plot each Companies figures onto the same plot
    plt.figure(figsize=(14, 8))
    for publisher in sales_per_year_publisher.columns:
        plt.plot(sales_per_year_publisher.index, sales_per_year_publisher[publisher], marker='o', label=publisher)

    plt.xlabel('Year')
    plt.ylabel('Global Sales (in millions)')
    plt.title('Global Sales per Year Comparison: Nintendo vs Electronic Arts vs Activision')
    plt.legend()
    plt.grid(True, linestyle='--', linewidth=0.5)
    plt.tight_layout()
    plt.show()
    sales_per_year_publisher
plot_top_three_publishers()
```
